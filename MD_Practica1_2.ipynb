# SOLUTION:

```python
import numpy as np

#Como nos pide "helper functions", he supuesto que se refeire a funciones para cada operacion, con errores de cada operacion
def suma(a, b):
    if a.shape != b.shape:
        raise ValueError(f"Incompatible shapes for addition: {a.shape} and {b.shape}") #El error de la suma
    return a + b #Devuelve la suma

def resta(a, b):
    if a.shape != b.shape:
        raise ValueError(f"Incompatible shapes for subtraction: {a.shape} and {b.shape}") #El error de la resta
    return a - b #Devuelve la resta

def multiplicacion_matricial(a, b):
    if a.shape[1] != b.shape[0]:
        raise ValueError(f"Incompatible shapes for matrix multiplication: {a.shape} and {b.shape}") #El error de la multiplicacion
    return a @ b #Devuelve la multiplicacion matricial

def concatenacion(a, b): 
    if a.shape[1] != b.shape[1]:#He supuesto que la concatenacion es con columnas iguales (habrán mas filas), ya que no se especifica en el PDF
        raise ValueError("Arrays must have the same number of columns to concatenate on axis 0") #El error de la concatenacion
    return np.concatenate((a, b), axis=0) #Devuelve la concatenacion de a y b

def transponer(a): #Una traspuesta no tiene error
    return a.T


def array_calculator(array1: np.ndarray, array2: np.ndarray, operation: str) -> np.ndarray:
    # Las operaciones de la calculadora hacia funciones de cada operacion
    if operation == '+':
        return suma(array1, array2)
    elif operation == '-':
        return resta(array1, array2)
    elif operation == '*':
        return multiplicacion_matricial(array1, array2)
    elif operation == 'concat':
        return concatenacion(array1, array2)
    elif operation == 'transpose':
        return transponer(array1)
    else:
        # Este es el error exacto que pide el PDF
        raise ValueError(f"Operation '{operation}' not supported")


def main():
    print("Accediendo a la calculadora de NumPy...")
    a = np.array([[0,1,2], #Nuestra matriz A, MODIFICABLE
                [3,4,5],
                [6,7,8]])
    print("Array A:")
    print(a)
    print()

    b = np.array([[0,1,2], #Nuestra matriz B, MODIFICABLE
                [3,4,5],
                [6,7,8]])
    print("Array B:")
    print(b)
    #lista_operaciones = ['+', '-', '*', 'concat', 'transpose'], para saber que operaciones se pueden hacer.

    signo = '*' #Nuestro signo (operacion que vamos a realizar), MODIFICABLE

    print()
    print(f"*** Ejecutando operación: {signo} ***")
    
    resultado = array_calculator(a, b, signo)
    print()
    print("Resultado:")
    print(resultado)

main()
```
