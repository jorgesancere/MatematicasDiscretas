## 6.6. Merging DataFrames

### 6.6.1. Merging two DataFrames

Podemos combinar dos dataframes en uno solo, gracias a la herramienta merge(), teniendo que mantener una de las columnas exactamente iguales entre dataframes (en mi caso la de nombre de los jugadores:

```python
import pandas as pd

# 1. Datos básicos
datos1 = {
    'Nombre': ['LeBron James', 'Michael Jordan', 'Kobe Bryant', 'Stephen Curry', 'Nikola Jokic'],
    'Equipo': ['Lakers', 'Bulls', 'Lakers', 'Warriors', 'Nuggets'],
    'Puntos': [25.7, 30.1, 25.0, 26.4, 29.7]
}

# 2. Datos de rendimiento físico (usando los mismos Nombres)
datos2 = {
    'Nombre': ['LeBron James', 'Michael Jordan', 'Kobe Bryant', 'Stephen Curry', 'Nikola Jokic'],
    'Rebotes': [7.3, 6.2, 5.2, 4.5, 13.7],
    'Asistencias': [8.3, 5.3, 4.7, 5.1, 11.7]
}

df1 = pd.DataFrame(datos1)
df2 = pd.DataFrame(datos2)

# Ahora el merge sí funcionará porque encuentra los mismos nombres en todos
df = pd.merge(df1, df2, on='Nombre')
df
```

### 6.6.2. Merging multiple DataFrames

Una vez lo hemos visto con dos, la herramienta merge tambien acepta combinar más de dos dataframes

```python
import pandas as pd

# 1. Datos básicos
datos1 = {
    'Nombre': ['LeBron James', 'Michael Jordan', 'Kobe Bryant', 'Stephen Curry', 'Nikola Jokic'],
    'Equipo': ['Lakers', 'Bulls', 'Lakers', 'Warriors', 'Nuggets'],
    'Puntos': [25.7, 30.1, 25.0, 26.4, 29.7]
}

# 2. Datos de rendimiento físico (usando los mismos Nombres)
datos2 = {
    'Nombre': ['LeBron James', 'Michael Jordan', 'Kobe Bryant', 'Stephen Curry', 'Nikola Jokic'],
    'Rebotes': [7.3, 6.2, 5.2, 4.5, 13.7],
    'Asistencias': [8.3, 5.3, 4.7, 5.1, 11.7]
}

# 3. Datos de relaciones (usando los mismos Nombres)
datos3 = {
    'Nombre': ['LeBron James', 'Michael Jordan', 'Kobe Bryant', 'Stephen Curry', 'Nikola Jokic'],
    'Amigos': [[8, 6], [10, 5], [4, 13], [5, 6], [1, 2]]
}

df1 = pd.DataFrame(datos1)
df2 = pd.DataFrame(datos2)
df3 = pd.DataFrame(datos3)

# Ahora el merge sí funcionará porque encuentra los mismos nombres en todos
df = pd.merge(df1, df2, on='Nombre')
df = pd.merge(df, df3, on='Nombre')

df
```
